```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data=pd.read_csv('StudentsPerformance.csv')

display(data.head(20))
```

```python
df=data['test preparation course']
df.value_counts().plot(kind='pie')
plt.axis('equal')
plt.title('Pie chart of various groups')
```

```python
data.lunch.value_counts().plot(kind='barh')
plt.ylabel('lunch')
plt.xlabel('Count')
plt.title('Lunch')
```

```python
df1=pd.DataFrame(data,columns=["gender","math score","reading score","writing score"])
malemask=df1["gender"]=="male"
dfmale = df1[malemask]

dfmaleavg=dfmale[["math score","reading score","writing score"]].mean()

dffemale = df1[~malemask]


dffemaleavg=dffemale[["math score","reading score","writing score"]].mean()

print(dfmaleavg)

print(dffemaleavg)
r1=[1,3,5]
r2=[2,4,6]
plt.bar(r1,dfmaleavg)
plt.bar(r2,dffemaleavg)
xt=("math","reading","writing")

# y_pos=np.arange(len(xt)*2)

plt.xticks([1.5,3.5,5.5],xt)
plt.legend(["male","female"],bbox_to_anchor=(1.25,0.9),loc="center right")
plt.xlabel("Areas of score",fontsize="15")
plt.ylabel("Score",fontsize="15")
plt.title("Comparision of male and female scores in subjects",fontsize="20")
plt.show()
```

```python
plt.figure(figsize=(10,10))
df2=pd.DataFrame(data, columns=['parental level of education','math score','reading score','writing score'])
df2['mean']=df2.loc[:,["math score","reading score","writing score"]].mean(axis=1)
plt.scatter(df2['parental level of education'],df2['mean'])
xt1=("bachelor's degree","some collage","masters degree","associates degree","high school","some high school")
plt.xticks([0.,1,2,3,4,5],xt1)

plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.hist(data['parental level of education'],edgecolor='red', bins=6)
xt1=("bachelor's degree","some collage","masters degree","associates degree","high school","some high school")
plt.xticks([0.35,1.25,2.05,2.95,3.8,4.6],xt1)
plt.xlabel('Various Degrees',fontsize="15")
plt.ylabel('Count',fontsize="15")
plt.title('Histogram For Parental Level Of Education',fontsize="15")
plt.show()
```
